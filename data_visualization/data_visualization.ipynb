```python
import geopandas
import pandas as pd
import plotly.express as px
```

```python
relative_logging_path = "log/20230129_13:01:12/rider.csv"
logging_path = "./../{}".format(relative_logging_path)
```

```python
df = pd.read_csv(logging_path)
```

```python
df.head()
```

```python
gdf = geopandas.GeoDataFrame(df, geometry=geopandas.points_from_xy(df.long, df.lat))
gdf.head()
```

```python
fig = px.scatter_mapbox(gdf, lat=gdf.geometry.y, lon=gdf.geometry.x, hover_name="id", hover_data=["action"],
                        animation_frame="time", animation_group="id", color="id",
                        color_discrete_sequence=["fuchsia"], zoom=10, height=300)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
from dash import Dash, dcc, html, Input, Output
import plotly.express as px

app = Dash(__name__)


app.layout = html.Div([
    html.H4('Multiorder visualization'),
    dcc.Graph(figure = fig, style={'width': '90vh', 'height': '90vh'}),
])


app.run_server(debug=True, use_reloader=False, port=8052)
```
